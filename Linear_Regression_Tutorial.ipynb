```python
import pandas as pd
```

# 2.  **Data Loading**

```python
data = pd.read_excel('Linear Regression_1D_data.xlsx')
```

```python
data
```

# 3. **Import matplotlib library**

- Used for data visualization in Python

- Helps plot graphs, charts, and trends

- Useful for understanding data patterns

```python
import matplotlib.pyplot as plt
```

# 4. **Visualize the data**

```python
plt.scatter(data['x'], data['y'], label='Data')
plt.xlabel('x', fontsize=14, fontweight='bold')
plt.ylabel('y', fontsize=14, fontweight='bold')
plt.title('x and y pairs', fontsize=14, fontweight='bold')
plt.legend()
```

# 5. **Define the model**

```python
def linear_1d(w_vec, x_vec=data['x']):
    w0 = w_vec[0]
    w1 = w_vec[1]

    pred = w0 + w1 * x_vec
    return pred
```

```python
w_arb_vec = [6, -2]
y_arb_pred = linear_1d(w_arb_vec, x_vec=data['x'])
```

# 6. **Plot the arbitrary model**

```python
plt.scatter(data['x'], data['y'], label = 'data')
plt.xlabel('x', fontsize=14, fontweight='bold')
plt.ylabel('y', fontsize=14, fontweight='bold')
plt.title('x and y pairs', fontsize=14, fontweight='bold')

plt.plot(data['x'], y_arb_pred, c='r', label= 'Arbitrary Line')
plt.legend()
```

```python
error = data['y'] - y_arb_pred
error
```

# 7. **Import Numpy library**

- Provides efficient numerical computation

- Supports multi-dimensional arrays and matrices

- Enables fast mathematical and linear algebra operations

```python
import numpy as np
```

# 8. **Compute Mean Square Error**

```python
mse_arb = np.mean(np.square(error))
mse_arb
```

```python
print('Mean Square Error: %.2f units' %mse_arb)
```

#9. **Compute optimal model**

```python
x_avg = np.mean(data['x'])
y_avg = np.mean(data['y'])

w1_num = np.sum((data['x'] - x_avg) * (data['y'] - y_avg))
w1_den = np.sum(np.square(data['x']-x_avg))
w1 =w1_num/ w1_den

w0 = y_avg - w1 * x_avg
```

```python
w0, w1
```

```python
print('w0: %.2f' %w0)
print('w1: %.2f' %w1)
```

```python
w_vec = [w0, w1]
y_pred = linear_1d(w_vec, x_vec=data['x'])
```

# 10. **Visualize the optimal line**

```python
plt.scatter(data['x'], data['y'], label = 'data')
plt.xlabel('x')
plt.xlabel('y')


plt.plot(data['x'], y_arb_pred, c='r', label='Arbitrary Line')

plt.plot(data['x'], y_pred, c='b', label= 'Optimal Fitted Line')

plt.legend()
```

# 11. **Compute MSE on optimal model**

```python
mse_optimal = np.mean(np.square(data['y']- y_pred))
```

```python
print('Mean square error: %.2f' %mse_optimal)
print('Mean square error: %.2f' %mse_arb)
```

# 12. **Make predictions for new data**

```python
new_x = np.array([0.2 , 1.2])
new_y_pred = linear_1d(w_vec, x_vec=new_x)
```

```python
new_y_pred
```

```python
plt.scatter(data['x'], data['y'], label = 'data')
plt.xlabel('x')
plt.xlabel('y')

plt.plot(data['x'], y_arb_pred, c='r', label= 'Arbitrary Line')

plt.plot(data['x'], y_pred, c='b', label= 'Optimal Fitted Line')

plt.scatter(new_x, new_y_pred, c='m', label= 'New Predictions')

plt.legend()
```

# 14. **Define the variables**

```python
data= pd.read_excel('Linear Regression_1D_data.xlsx')
```

```python
data
```

```python
X = data[['x']]
y = data['y']
```

```python
X.shape
```

# 15. **Define the model**

```python
LR_model = LinearRegression()

#training
LR_model.fit(X,y)

#prediction
LR_y_pred = LR_model.predict(X)
```

# 15. **Plot the regression line**

```python
plt.scatter(data['x'], data['y'], label='data')
plt.xlabel('x', fontsize=14, fontweight='bold')
plt.ylabel('y', fontsize=14, fontweight='bold')
plt.title('x and y pairs', fontsize=14, fontweight='bold')

plt.plot( X, LR_y_pred, c='r', label ='Library fitted line')

plt.legend
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_LR = mean_squared_error(data['y'], LR_y_pred)
mse_LR
```

```python
print('MSE: %.2f' %mse_LR)
print('MSE: %.2f' %mse_optimal)
```
